```python
from pathlib import Path
```

```python
import numpy as np
import xarray as xr
```

```python
import cartopy.crs as ccrs
```

```python
DIR = Path('../demo_20210908/2020/202007/20200701/20200701T1200/')
```

# Solar Zenith

```python
f = next(DIR.glob('*solar_zenith*'))
ds = xr.open_dataset(f)
dt = ds.start_time.to_pandas().item()
```

```python
figure(figsize=(7,3))
ax = axes(projection=ccrs.PlateCarree())
ds.solar_zenith_angle.squeeze()[::10,::10].plot(cmap='RdBu', vmin=0, vmax=180)
title(dt.strftime('Solar Zenith Angle %Y-%m-%d %H:%M'))
ax.coastlines()
```

# Solar Azimuth

```python
f = next(DIR.glob('*solar_azimuth*'))
ds = xr.open_dataset(f)
dt = ds.start_time.to_pandas().item()
```

```python
figure(figsize=(7,3))
ax = axes(projection=ccrs.PlateCarree())
ds.solar_azimuth_angle.squeeze()[::10,::10].plot(cmap='hsv', vmin=0, vmax=360)
title(dt.strftime('Solar Azimuth Angle (%Y-%m-%d %H:%M)'))
ax.coastlines()
```

# Satellite Zenith

```python
f = next(DIR.glob('*satellite_zenith*'))
ds = xr.open_dataset(f)
dt = ds.start_time.to_pandas().item()
```

```python
figure(figsize=(7,3))
ax = axes(projection=ccrs.PlateCarree())
ds.satellite_zenith[0,0,::10,::10].plot(cmap='RdBu', vmin=0, vmax=180)
title(dt.strftime('Satellite Zenith Angle Layer=0 (%Y-%m-%d %H:%M)'))
ax.coastlines()
```

# Satellite Azimuth

```python
f = next(DIR.glob('*satellite_azimuth*'))
ds = xr.open_dataset(f)
dt = ds.start_time.to_pandas().item()
```

```python
figure(figsize=(7,3))
ax = axes(projection=ccrs.PlateCarree())
ds.satellite_azimuth[0,0,::10,::10].plot(cmap='hsv', vmin=-180, vmax=180)
title(dt.strftime('Satellite Azimuth Angle Layer=0 %Y-%m-%d %H:%M'))
ax.coastlines()
```
